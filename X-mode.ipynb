```python
casino_loc = (39.959083, 116.418655) # Our dummy location
```

```python
geopy.distance.EARTH_RADIUS = 6371.00877141 # Change the radius of the earth's accuracy
```

```python
df = pd.read_csv(r'C:\Users\Natarajan\Desktop\big_set.csv') # Read the dataset
df['loc_tuple'] = list(zip(round(df['lat'],10), round(df['long'],10))) # Create the location tuple
# Apply the Vincenty function to each location
%time df['vincenty'] = df['loc_tuple'].apply(lambda x : geopy.distance.vincenty(casino_loc,x).m)
df.head()
```

```python
df['accuracy'] = 10 * np.random.rand(len(df)) # Simulate the accuracy variable
%time df['corrected'] = df['vincenty'] + df['accuracy'] # Create the corrected distance
df.head()
```
